```python
import pandas as pd
import numpy as np
import types
import pandas as pd

url_html='https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&direction=next&oldid=942655364'

df = pd.read_html(url_html)

df_postcodes=df[0]

print("imported dataframe has",df_postcodes['Postcode'].count(), "postcodes entries")

indexNames = df_postcodes[(df_postcodes['Borough'] == "Not assigned") & (df_postcodes['Neighbourhood'] == "Not assigned")].index

df_postcodes.drop(index=indexNames, inplace=True)
df_postcodes.groupby(by=['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()


```
